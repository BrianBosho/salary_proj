```python
df = df[df['Salary Estimate'] != "-1"]
```

```python
df.shape
```

```python
df
```

```python
df['hourly']= df['Salary Estimate'].apply(lambda x: 1 if '/hr' in x.lower() else 0)
```

```python
df['yearly']= df['Salary Estimate'].apply(lambda x: 1 if '/yr' in x.lower() else 0)
```

```python
df.head()
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('/')[0])
df['Salary'] = salary
```

```python
df.head()
```

```python
df['Salary'] = df['Salary'].apply(lambda x: x.replace('$', ''))
df.head()
```

```python
df.shape
```

```python
df['age'] = df['Founded'].apply(lambda x:x if x ==-1 else 2022-x)
# df['Founded'].value_counts()
df.head()
```

```python
df.sample(10)
```

```python
df['Location'].value_counts()
```

```python
# df['state']= df['Location'].apply(lambda x: x if x=='Remote' else x.split(',')[1])
def get_state(x):
    location = x.split(',')
    if len(location)>1:
        state = location[1]
    else:
        state = location[0]
    return state

df['state']= df['Location'].apply(get_state)
df.sample(5)
```
