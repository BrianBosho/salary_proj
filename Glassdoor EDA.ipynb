```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('500jobs_cleaned.csv')
df.head()
```

```python
df.columns
```

```python
df['hourly'].value_counts()
```

```python
df['Salary'] = df['Salary'].apply(lambda x: x.split('.')[0])
df['Salary'] = df['Salary'].apply(lambda x: x.split(',')[0])
df['Salary'] = df['Salary'].astype('int')
```

```python
df.head()
```

```python
df['Salary'] = df.apply(lambda x: x.Salary*2000 if x.hourly==1 else x.Salary*1000, axis=1)
df.head()
```

```python
df_cat = df[['Job Title','state','Size','Type of ownership', 'Industry', 'Sector', 'Revenue']]
df_cat.head()
```

```python
df_cat = df_cat[df_cat.Size != "-1"]
# df_cat.drop([df_cat.Size =="-1"], inplace=True, axis=)
df_cat.head()
```

```python
df_cat.describe()
```

```python
for column in df_cat.columns:  
    counts = df_cat[column].value_counts()
    chart = sns.barplot(x=counts.index ,y=counts)
    chart.set_xticklabels(chart.get_xticklabels(), 
                              rotation=90, 
                              horizontalalignment='right')
    plt.show()
```
